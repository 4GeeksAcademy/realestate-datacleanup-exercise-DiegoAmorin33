#### Ejercicio 11. Analiza la relación entre la superficie y el precio de las casas. (★★☆)

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for poblacion in cinturon_sur['level5'].unique():
    precios = cinturon_sur[cinturon_sur['level5'] == poblacion]['price']
    precios_norm = (precios - precios.min()) / (precios.max() - precios.min())
    plt.hist(precios_norm, alpha=0.5, label=poblacion)

plt.title('Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Getafe presenta precios más homogéneos, concentrados en valores intermedios. Leganés también muestra precios agrupados en rangos medios. Fuenlabrada y Alcorcón tienen distribuciones más dispersas, con precios repartidos en todo el espectro, siendo Fuenlabrada la que mayor variedad de precios presenta.

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, basemaps, CircleMarker

map = Map(center=(40.35, -3.8), zoom=11.5, 
          basemap=basemaps.OpenStreetMap.Mapnik)

colores = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue', 
    'Getafe': 'green',
    'Alcorcón': 'orange'
}

for idx, fila in cinturon_sur.iterrows():
    lat = float(str(fila['latitude']).replace(',', '.'))
    lon = float(str(fila['longitude']).replace(',', '.'))
    
    marcador = CircleMarker(
        location=(lat, lon),
        radius=5,
        color=colores[fila['level5']],
        fill_color=colores[fila['level5']]
    )
    map.add_layer(marcador)

map
```
